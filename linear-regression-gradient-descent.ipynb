```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Defining the cost function for a given set of data points (x, y) and line parameters (w, b)

def cost_function(w, b, x, y):
    n = len(x)
    error = np.sum(((w * x + b) - y)**2)
    total_error = error / (2*n) 
    return total_error
```

```python
# Step 2: Computing the gradients with respect to w and b

def gradients(w, b, x, y):
    n = len(x)
    dw = (1/n) * np.sum(x * ((w * x + b) - y))
    db = (1/n) * np.sum((w * x + b) - y)
    return dw, db
```

```python
# Step 3: Updating the parameters (w and b) using gradient descent

def gradient_descent(w, b, x, y, learning_rate, iterations):
    for _ in range(iterations):
        dw, db = gradients(w, b, x, y)
        w -= learning_rate * dw
        b -= learning_rate * db
    return w, b
```

```python
# Step 4: Testing the algorithm on a dataset assuming x and y are NumPy arrays containing the data points

x = np.array([1.1, 2, 2.9, 4, 5.05])
y = np.array([3, 4.05, 5, 5.9, 7])
```

```python
# Initializing w and b to any values

w_init, b_init = 0, 0
```

```python
# Setting the learning rate and number of iterations

learning_rate = 0.01
iterations = 1000
```

```python
# Performing gradient descent to find the best fit line

w_optimal, b_optimal = gradient_descent(w_init, b_init, x, y, learning_rate, iterations)
```

```python
# Printing the results

print(f"Optimal w: {w_optimal}, Optimal b: {b_optimal}")
```

```python
# Step 6: Visualizing the cost function

def plot_cost_function(x, y, w_optimal, b_optimal):
    w_values = np.linspace(w_optimal - 2, w_optimal + 2, 100)
    b_values = np.linspace(b_optimal - 2, b_optimal + 2, 100)
    
    W, B = np.meshgrid(w_values, b_values)
    cost = np.array([cost_function(w, b, x, y) for w, b in zip(np.ravel(W), np.ravel(B))])
    Cost = cost.reshape(W.shape)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    ax.plot_surface(W, B, Cost, cmap='viridis', alpha=0.8)
    ax.scatter(w_optimal, b_optimal, cost_function(w_optimal, b_optimal, x, y), c='red', marker='o', s=100)
    
    ax.set_xlabel('w')
    ax.set_ylabel('b')
    ax.set_zlabel('Cost')

    plt.show()
```

```python
# Calling the function to plot the cost function

plot_cost_function(x, y, w_optimal, b_optimal)
```
